##Question 1 : Word Count
##Mapper Function:

##Input
  - Key
  : **none**
    - The mapper function does not require the key argument as input. The key value mapping is done inside the mapper function itself.
  - Value
  : **line**
     - The line argument is passed by the MRJob framework by reading the  file line by line and passes each line as the **line** argument to the mapper.

##Output
  - Key
  : **word**

    - Words of the file are converted to lowercase for generalisaton and treated as a key corresponding to which a value in the form of list will be mapped. All the distinct words will be stored as key in **word**.

  - Value
  : List of **1**s based on the number of times a word has occured.
    
    - Value of **1** is appended in the list mapped to the key (word which is encountered) every time a particular word is encountered.



##Reducer Function:

##Input
  - Key
    : **word**

    - **word** is a key to which **counts** is mapped. It consists of all the words that have occured in the file.
  - Value
    : **counts**

    - **counts** contains a list mapped to a particular word denoting how many times that word has occured. The list consists of all **1**s.

##Output
  - Key
  : **word**
    - **word** is a key to which **word count** is mapped. It consists of all the words that have occured in the file.
  - Value
  : **word count**
    - **Word count** returns the number of times a particular word has occured in the file by summing the number of 1s from the list **counts** given as input.

### Question 1: Compute department wise total salary.
### $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$SELECT dno, sum(salary) from employee group by dno;  

```python
%%file empsum.py
from mrjob.job import MRJob
import re

class EmpSum(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      salary = int(record[3])
      yield dno, salary

    def reducer(self, dno, salaries):
        yield dno, sum(salaries)

if __name__ == '__main__':
     EmpSum.run()
```

###Question 2: Compute department wise maximum salary among employees from Massachusetts (’MA’).
###$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$SELECT dno, max(salary) from employee where state='MA' group by dno;

```python
%%file empmaxsal.py
from mrjob.job import MRJob
import re

class EmpMaxSal(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      salary = int(record[3])
      state = record[4]
      if(state == 'MA'):
        yield dno, salary

    def reducer(self, dno, salaries):
        yield dno, max(salaries)

if __name__ == '__main__':
     EmpMaxSal.run()
```

###Question 3: Compute department wise average salary.
###$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$SELECT dno, avg(salary) from employee group by dno;  

```python
%%file empavg.py
from mrjob.job import MRJob
import re

class EmpAvg(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      salary = int(record[3])
      yield dno, [salary , 1]

    def reducer(self, dno, salaries):
        '''salaries = list(salaries)'''
        sum = 0
        count = 0
        for salary in salaries:
          sum+=salary[0]
          count+=salary[1]
        avg = int(round(sum/count,0))
        yield dno, avg

if __name__ == '__main__':
     EmpAvg.run()
```

###Question 5: Compute total number of male and female employees for each department.
###$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$SELECT dno, gender, count(empno) from employee group by dno, gender;
  

```python
%%file gendercount.py
from mrjob.job import MRJob
import re

class GendCount(MRJob):
    def mapper(self, key, line):
      record = re.split(',', line)
      dno = record[2]
      gender = record[5]
      if(gender == 'F'):
        yield [dno,'F'], 1
      else:
        yield [dno,'M'], 1


    def reducer(self, dno_gender, count):
        yield dno_gender, sum(count)

if __name__ == '__main__':
     GendCount.run()
```

###Question 6: Computes following summaries on a Monthly Basis –
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$(a) Total number of requests.  
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$(b) Total download size (in MegaBytes)

```python
%%file monthly_summ.py
from mrjob.job import MRJob

class MonthlySumm(MRJob):

  def mapper(self, key, line):
    request = line.split(' ')
    date = request[3]
    date = date.split('/')
    month = date[1]
    year = date[2][:4]
    year_month = year + '-' + month
    data_size = request[9]
    if(data_size == '-'):
      data_size = 0
      yield year_month, [1,data_size]
    else:
      data_size = int(data_size)
      yield year_month, [1,data_size]


  def reducer(self, key, values):
      request_count = 0
      total_download_size = 0
      for value in values:
        request_count += value[0]
        total_download_size += value[1]

      total_download_size = round(total_download_size/(1024*1024),2)
      yield key, [request_count, total_download_size]

if __name__ == '__main__':
    MonthlySumm.run()
```
